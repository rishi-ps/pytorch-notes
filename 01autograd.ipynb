so basically, nn (neural network)(collection of nested functions defined by "parameters") training happens in two steps:

1. **Forward propagation**: runs input data through the collection of nested functions.

2. **Backward propagation**: by traversing backwards, nn adjusts the parameters proportionate to the error (from the correct output) by collecting the derivatives of the error with respect to the parameters of the functions.

an example using resnet for some reason:

*"We create a random data tensor to represent a single image with 3 channels, and height & width of 64, and its corresponding label initialized to some random values. Label in pretrained models has shape (1,1000)"*

```python
import torch
from torchvision.models import resnet18, ResNet18_Weights
model = resnet18(weights=ResNet18_Weights.DEFAULT)
data = torch.rand(1, 3, 64, 64)
labels = torch.rand(1, 1000)  # use llms to sort things out if you are getting dependency errors
```

```python
print(f"data: {data.dtype}, labels: {labels.dtype}")
```

```python
prediction = model(data)
prediction.shape, prediction.dtype
```

```python
loss = (prediction - labels)
loss.shape, loss.dtype
```

```python
loss = loss.sum()
loss.shape, loss.dtype, loss
```

```python
loss.backward()
```

```python
optim = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)
```

```python
optim.step() # gradient descent
```
